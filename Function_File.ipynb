```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data=pd.read_excel("D:DataSets/Airline.xlsx")
df=data
df.head()
```

```python
def date_time(col):
    df[col]=pd.to_datetime(df[col])
```

```python
def extract_hr_min(col):
    df[col+'_hour']=df[col].dt.hour
    df[col+'_min']=df[col].dt.minute
```

```python
def one_hot_encoding(data,col):
    print(data[col].value_counts())
    plt.figure(figsize=(15,5))
    print(sns.boxplot(x=col, y='Price', data=df.sort_values('Price', ascending=False)))
```

```python
def plot(df1,col):
    plt.figure(figsize=(30,20))
    fig,(ax1,ax2)=plt.subplots(2,1)
    sns.distplot(df1[col], ax=ax1)
    sns.boxplot(df1[col], ax=ax2)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
def predict(model,x_train,y_train,x_test,y_test):
    model_1=model.fit(x_train,y_train)
    
    TS=model_1.score(x_train,y_train)
    print('Training Score: {}'.format(TS))
    pred=model_1.predict(x_test)
    print("Predictions are {}".format(pred))
    print("\n")
    
    score=r2_score(y_test,pred)
    print("Score: {}".format(score))
    
    MAE=mean_absolute_error(y_test,pred)
    print("MAE : {}".format(MAE))
    
    MSE=mean_squared_error(y_test,pred)
    print("MSE : {}".format(MSE))
    
    RMSE=np.sqrt(mean_absolute_error(y_test,pred))
    print("RMSE : {}".format(RMSE))
    sns.distplot(y_test-pred)
    
    return(TS,score,MAE,MSE,RMSE)
```

```python
from sklearn.model_selection import RandomizedSearchCV

def randomcv(RFE,rabdm_grid,xtrain,ytrain,x_test,y_test):
    random_cv=RandomizedSearchCV(estimator=RFE, param_distributions=rabdm_grid, cv=3, verbose=2, n_jobs=-1)
    random_cv.fit(xtrain,ytrain)
    print("The best parameters are ", random_cv.best_params_)
    pred=random_cv.predict(x_test)
    
    score=r2_score(y_test,pred)
    print("Score {}".format(score))
    
    print("MAE : {}".format(mean_absolute_error(y_test,pred)))
    print("MSE : {}".format(mean_squared_error(y_test,pred)))
    print("RMSE : {}".format(np.sqrt(mean_absolute_error(y_test,pred))))
    sns.distplot(y_test-pred)
    
```
